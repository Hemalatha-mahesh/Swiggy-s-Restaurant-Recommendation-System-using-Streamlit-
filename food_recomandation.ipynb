```python
# label encoder for encoding cuisine and city column
from sklearn.preprocessing import LabelEncoder

# Create LabelEncoder objects
le_city = LabelEncoder()
le_cuisine = LabelEncoder()

#  Encode city column 
cleaned_data["city_encoded"] = le_city.fit_transform(cleaned_data["city"])

#  Encode cuisine column 
cleaned_data["cuisine_encoded"] = le_cuisine.fit_transform(cleaned_data["cuisine"])

#  Print encoded values 
print("Encoded columns:")
print(cleaned_data[["city", "city_encoded", "cuisine", "cuisine_encoded"]].head())

#  Mapping for city 
print("\nCity Encoding Mapping:")
city_mapping = dict(zip(le_city.classes_, le_city.transform(le_city.classes_)))
print(city_mapping)

# Mapping for cuisine 
print("\nCuisine Encoding Mapping:")
cuisine_mapping = dict(zip(le_cuisine.classes_, le_cuisine.transform(le_cuisine.classes_)))
print(cuisine_mapping)

```

```python
# Indicies matching -> cleaned data and encoded data

import pandas as pd

# Load the encoded data and cleaned data
encoded_data = pd.read_csv('D:/Swiggy/encoded_data.csv', index_col=0)
cleaned_data = pd.read_csv('D:/Swiggy/cleaned_data.csv', index_col=0)

# Align the DataFrames by common indices
encoded_data_aligned = encoded_data.loc[encoded_data.index.isin(cleaned_data.index)]
cleaned_data_aligned = cleaned_data.loc[cleaned_data.index.isin(encoded_data.index)]

# Verify the Alignment if the indices are now aligned
print(encoded_data_aligned.index.equals(cleaned_data_aligned.index))
```

```python
#clustering
from sklearn.cluster import KMeans

# Select numeric features
features = encoded_data[[
    "rating", 
    "rating_count",
    "cost",
    "city",
    "cuisine"
]]

# KMeans clustering
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
encoded_data["cluster"] = kmeans.fit_predict(features)

print("Clustering completed.")
print(encoded_data.head())
```

```python
#Top 5 recommendation
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

#Load Encoded Data
encoded_data = pd.read_csv(r"D:/Swiggy/encoded_data.csv", index_col=0)

# KMeans Clustering Function
def kmeans_clustering(data, n_clusters=5):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    data = data.copy()
    data["Cluster"] = kmeans.fit_predict(data)
    return data, kmeans

# Perform clustering
encoded_data_with_clusters, kmeans_model = kmeans_clustering(encoded_data)

# Load Original Cleaned Data
cleaned_data = pd.read_csv(r"D:/Swiggy/cleaned_data.csv", index_col=0)

# Add cluster labels to original data
cleaned_data["Cluster"] = encoded_data_with_clusters["Cluster"]

# Select Restaurant by Position
restaurant_position = 0    

# Get the cluster of this restaurant
restaurant_cluster = cleaned_data.iloc[restaurant_position]["Cluster"]

# Get Similar Restaurants

# All restaurants in same cluster
same_cluster_indices = cleaned_data[cleaned_data["Cluster"] == restaurant_cluster].index.tolist()

# Remove the selected restaurant itself
selected_index = cleaned_data.index[restaurant_position]
if selected_index in same_cluster_indices:
    same_cluster_indices.remove(selected_index)

# Pick top 5 restaurants
top_5_recommendations = cleaned_data.loc[same_cluster_indices[:5]]

# Display Output
print("\nSelected Restaurant:")
print(cleaned_data.iloc[restaurant_position])

print("\nTop 5 Recommended Restaurants:")
print(top_5_recommendations)
```
